### IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### LOADING THE DATASET

```python
df = pd.read_csv('CarPrice.csv')
df.head()
```

### DATA PREPROCESSING

```python
df.shape
```

```python
df.columns
```

```python
df1 = df.drop(['car_ID','symboling','aspiration','doornumber','enginelocation','curbweight','boreratio','compressionratio'],axis = 'columns')
df1.head()
```

```python
df1.shape
```

### LABEL ENCODER

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
df1['CarName'] = le.fit_transform(df1['CarName'])
df1['fueltype'] = le.fit_transform(df1['fueltype'])
df1['carbody'] = le.fit_transform(df1['carbody'])
df1['drivewheel'] = le.fit_transform(df1['drivewheel'])
df1['drivewheel'] = le.fit_transform(df1['drivewheel'])
df1['enginetype'] = le.fit_transform(df1['enginetype'])
df1['cylindernumber'] = le.fit_transform(df1['cylindernumber'])
df1['fuelsystem'] = le.fit_transform(df1['fuelsystem'])
```

```python
df1
```

### SEPARATING DATA LABELS FOR TRAIN TEST SPLIT

```python
x = df1.drop('price',axis = 'columns')
y = df1['price']
```

### TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25)
```

```python
len(x_train)
```

```python
len(y_test)
```

### MODEL BUILDING

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict(x_test)
```
